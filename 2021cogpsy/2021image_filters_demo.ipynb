```python
from scipy import ndimage, misc
import matplotlib.pyplot as plt
%matplotlib inline
import japanize_matplotlib
from skimage.color import rgb2gray

fig = plt.figure(figsize=((16,8)))
#plt.gray()  # show the filtered result in grayscale

ax1 = fig.add_subplot(241)  # left side
ax2 = fig.add_subplot(242)
ax3 = fig.add_subplot(243)
ax4 = fig.add_subplot(244)
ax21 = fig.add_subplot(245)
ax22 = fig.add_subplot(246)
ax23 = fig.add_subplot(247)

#img = misc.ascent()
img = rgb2gray(plt.imread('/Users/asakawa/Pictures/DSC01323.jpg'))
sobel_a = ndimage.sobel(img)
prewitt_a = ndimage.prewitt(img)
laplace_a = ndimage.laplace(img)
gaussian_a = ndimage.gaussian_filter(img, sigma=5)
maximum_a = ndimage.maximum_filter(img, size=20)
lapgap = ndimage.laplace(gaussian_a)
#conv = ndimage.convolve(img, ,

ax1.imshow(img, cmap='gray'); ax1.set_xlabel('')
ax2.imshow(sobel_a, cmap='gray'); ax2.set_xlabel('')
ax3.imshow(prewitt_a, cmap='gray'); ax3.set_xlabel('')
ax4.imshow(maximum_a, cmap='gray'); ax4.set_xlabel('')
ax21.imshow(gaussian_a, cmap='gray'); ax21.set_xlabel('')
ax22.imshow(laplace_a, cmap='gray'); ax22.set_xlabel('')
ax23.imshow(lapgap, cmap='gray'); ax23.set_xlabel('')
plt.show()
```
