## 2. Pythonの基礎

### 2.1 変数とデータ型

````python
# 整数（int）
x = 10
print(f"x = {x}, 型: {type(x)}")

# 浮動小数点数（float）
y = 3.14
print(f"y = {y}, 型: {type(y)}")

# 文字列（str）
name = "人工知能"
print(f"name = {name}, 型: {type(name)}")

# ブール値（bool）
flag = True
print(f"flag = {flag}, 型: {type(flag)}")
````

### 2.2 リスト（list）

複数のデータをまとめて扱うためのデータ構造

````python
# リストの作成
scores = [85, 92, 78, 95, 88]
print("成績リスト:", scores)

# 要素へのアクセス（インデックスは0から始まる）
print("1番目の成績:", scores[0])
print("3番目の成績:", scores[2])

# スライス（範囲指定）
print("最初の3つ:", scores[0:3])
print("最後の2つ:", scores[-2:])

# リストの長さ
print("データ数:", len(scores))
````

## 3. NumPy による数値計算

NumPy は科学計算のための基本ライブラリです。

### 3.1 NumPy のインポート

````python
# NumPyのインポート（慣例的にnpという名前で使用）
import numpy as np

print("NumPy version:", np.__version__)
````

### 3.2 配列の作成

````python
# リストから配列を作成
arr1 = np.array([1, 2, 3, 4, 5])
print("配列1:", arr1)
print("型:", type(arr1))
print("形状:", arr1.shape)

# 2次元配列
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])
print("\n配列2:")
print(arr2)
print("形状:", arr2.shape)  # (行数, 列数)
````

### 3.3 便利な配列生成関数

````python
# ゼロで埋められた配列
zeros = np.zeros(5)
print("zeros:", zeros)

# 1で埋められた配列
ones = np.ones(5)
print("ones:", ones)

# 等間隔の数列（0から10まで6個）
linspace = np.linspace(0, 10, 6)
print("linspace:", linspace)

# ランダムな値（0〜1の範囲）
random = np.random.rand(5)
print("random:", random)
````

### 3.4 配列の演算

````python
# 配列の作成
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

# 要素ごとの演算
print("a + b =", a + b)
print("a * b =", a * b)
print("a ** 2 =", a ** 2)  # 2乗

# 統計量
print("\n平均:", np.mean(a))
print("標準偏差:", np.std(a))
print("最大値:", np.max(a))
print("最小値:", np.min(a))
````

## 5. データ・モデル・評価の実例

### 5.1 データの準備

````python
# 簡単な例：学習時間と試験成績の関係

# データ（学習時間と成績）
study_hours = np.array([1, 2, 3, 4, 5, 6, 7, 8])
test_scores = np.array([50, 55, 60, 65, 70, 75, 80, 85])

print("学習時間:", study_hours)
print("試験成績:", test_scores)

# データの可視化
plt.figure(figsize=(10, 6))
plt.scatter(study_hours, test_scores, s=100, alpha=0.7, label='実際のデータ')
plt.xlabel('学習時間（時間）', fontsize=12)
plt.ylabel('試験成績（点）', fontsize=12)
plt.title('学習時間と試験成績の関係', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
````

### 5.2 モデル：線形回帰

scikit-learnを使って最もシンプルなモデルを作成します。

````python
# scikit-learnから線形回帰モデルをインポート
from sklearn.linear_model import LinearRegression

# データの形状を整える（scikit-learnは2次元配列を要求）
X = study_hours.reshape(-1, 1)  # (-1, 1)は「行数は自動、列数は1」の意味
y = test_scores

# モデルの作成と学習
model = LinearRegression()
model.fit(X, y)

# 学習されたパラメータ
print(f"傾き（係数）: {model.coef_[0]:.2f}")
print(f"切片: {model.intercept_:.2f}")
print(f"\n式: y = {model.coef_[0]:.2f} * x + {model.intercept_:.2f}")
````

### 5.3 予測と可視化

````python
# 予測値の計算
y_pred = model.predict(X)

# 実際のデータと予測結果を可視化
plt.figure(figsize=(10, 6))
plt.scatter(study_hours, test_scores, s=100, alpha=0.7, label='実際のデータ')
plt.plot(study_hours, y_pred, 'r-', linewidth=2, label='予測（モデル）')
plt.xlabel('学習時間（時間）', fontsize=12)
plt.ylabel('試験成績（点）', fontsize=12)
plt.title('線形回帰による予測', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# 新しいデータで予測
new_hours = np.array([[9], [10]])  # 9時間、10時間勉強した場合
predicted_scores = model.predict(new_hours)

print("\n予測結果:")
print(f"9時間勉強 → 予測成績: {predicted_scores[0]:.1f}点")
print(f"10時間勉強 → 予測成績: {predicted_scores[1]:.1f}点")
````

### 5.4 評価：決定係数（$R^2$）

モデルがデータをどれだけ説明できているかを測る指標

````python
# 決定係数の計算
from sklearn.metrics import r2_score, mean_squared_error

r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print(f"決定係数（R²）: {r2:.4f}")
print(f"平均二乗誤差（MSE）: {mse:.4f}")
print(f"\nR² = 1.0 に近いほど良いモデル")
print(f"今回のR² = {r2:.4f} は、モデルがデータをよく説明していることを示す")
````
